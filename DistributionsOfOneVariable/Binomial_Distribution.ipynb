## 1. Bernoulli Distribution

The simplest case of a univariate distribution, and also the basis of the binomial distribution, is the *Bernoulli distribution*. Bernoulli distribution has only two states.

**Example:** Coin flip

State 1: Heads, $p_{heads} = 0.5$

State 2: Tails, $p_{tails} = 0.5$

$$ p_{heads} + p_{tails} = 1 $$




```python
# Python code

from scipy import stats
p = 0.5

# frozen distribution
bernoulliDist = stats.bernoulli(p)

# Probabilities of the Bernoulli distribution
print("p_tails:", bernoulliDist.pmf(0))
print("p_heads:", bernoulliDist.pmf(1))

# generate random variable from Bernoulli distribution
trials = bernoulliDist.rvs(10)
print("\nRandom variables from the Bernoulli distribution:\n", trials)
```

## 2. Binomial Distribution

In general, the binomial distribution is associated with the question: 

*Out of a given (fixed) number of trials, how many will succeed?*

**More examples:**

- Out of ten tosses, how many times will the coin land heads up?

- From the children born in a given hospital on a given day, how many of them will be girls?

- How many students in a given classroom will have blue eyes?

Suppose, we conduct $n$ repeated experiments where the probability of success is given by the parameter $p$
and add up the number of successes. This number of successes is represented by the random variable $X$. The value of $X$ is between 0 and $n$. $X$ is binomially distributed: $X \in B(n,p)$.

$P(X=k)$ is the probability of $k$ successes in a binomial experiment.

$$
p_k = P(X=k) = \begin{cases}
  C_{n}^{k} * p^k * (1 - p)^{n-k}, 0 \le k \le n, 0 \le p \le 1, n \in \mathbb{N} \\    
  0 \space otherwise
\end{cases}
$$

where $C_{n}^{k} = \frac{n!}{k!(n-k)!}$

$$ \sum_{i=0}^{i=n} p_i = 1 $$

```python
# Python code

from scipy import stats 

N = 5
p = 0.5

# Define Binomial distribution
binomDist = stats.binom(N, p)

# Probabilities
bProb = binomDist.pmf(range(N+1))

print("\nBinomial Distribution", "\nValues:", list(range(N+1)), "\nProbabilities:", bProb)

```

The chances that in 5 coin tosses the heads never come up is about 3%.

```python
# Binomial Distribution functions

## p=0.5, n=25
N1 = 25
p1 = 0.5
binomDist1 = stats.binom(N1, p1)
x1 = list(range(N1+1))
y1 = binomDist1.pmf(x1)

## p=0.5, n=50
N2 = 50
p2 = 0.5
binomDist2 = stats.binom(N2, p2)
x2 = list(range(N2+1))
y2 = binomDist2.pmf(x2)

## p=0.7, n=25
N3 = 25
p3 = 0.7
binomDist3 = stats.binom(N3, p3)
x3 = list(range(N3+1))
y3 = binomDist3.pmf(x3)
```

```python
# Q2: If we have 15 Austrian students in the classroom, 
# what is the chance of finding three, six, or ten students with blue eyes?

# Q3: If we have 15 Swedish students in the classroom, 
# what is the chance of finding three, six, or ten students with blue eyes?

# Q4: If we have 15 Lithuanian students in the classroom, 
# what is the chance of finding three, six, or ten students with blue eyes?

from scipy import stats 
N = 15
values = [3, 6, 10]

p_at = 0.37
binomDist_at = stats.binom(N, p_at)
print("\nAustria")
for val in values:
    print(f"{val} students out of {N} has blue eyes: {round(binomDist_at.pmf(val),6)}")

p_se = 0.72
binomDist_se = stats.binom(N, p_se)
print("\nSweden")
for val in values:
    print(f"{val} students out of {N} has blue eyes: {round(binomDist_se.pmf(val),6)}")

p_lt = 0.61
binomDist_lt = stats.binom(N, p_lt)
print("\nLithuania")
for val in values:
    print(f"{val} students out of {N} has blue eyes: {round(binomDist_lt.pmf(val),6)}")
```

## 5. Binomial Test

In this section we will solve a hypothesis testing problem. 

Suppose we have a board game that depends on the roll of a die and attaches special importance to rolling a 6.
In a particular game, the die was rolled 235 times, and the six came up 51 times.

Is the die fair? Is the proportion of six to come up is significantly higher than would be expected by chance,
on the null hypothesis of a fair die?

Plot the distribution of six to come up in 100 die rolls.

### Solution:

$
H0: \pi = 1/6 \\
H1: \pi \ne 1/6
$

Level of significance: $\alpha = 0.05$

We will use a p-value approach and we will use the binomial distribution and the binomial test in order to solve the problem.

$\pi = N * p = 235 * 1/6 = 39.17 \approx 39$

$p_{value} = P(X \le 27) + P(X \ge 51)$

### Manual computation of p-value

```python
# Manual computation of p-value
from math import sqrt

m = 51
N = 235
p = 1/6
alpha = 0.05

dist = stats.binom(N, p)

# p-value for the two-tailed test
p_value = 1 - sum(dist.pmf(range(28, 51)))

# Conclusion
if p_value < alpha:
    print(f"We reject H0. p-value = {p_value}. Conclusion: The die is unfair.")
else:
    print(f"We do not reject H0. p-value = {p_value}. Conclusion: We assume that the die is fair.")
```

### SciPy binomtest

```python
# We could also use the SciPy library function to perform the binomial test
from scipy import stats

test_result = stats.binomtest(51, n=235, p=1/6, alternative="two-sided")

p_value = test_result.pvalue

# Conclusion
if p_value < alpha:
    print(f"We reject H0. p-value = {p_value}. Conclusion: The die is unfair.")
else:
    print(f"We do not reject H0. p-value = {p_value}. Conclusion: We assume that the die is fair.")
```
